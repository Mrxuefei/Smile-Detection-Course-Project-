```python
import time
import random
import pandas as pd
from sklearn import svm, metrics
from sklearn.decomposition import PCA
import collections
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
data = pd.read_csv('../fer2013.csv')
print data.shape

for i in range(data.shape[0]):
#     if data['emotion'][i] == 3:
#         data['emotion'][i] = 1
#     else:
#         data['emotion'][i] = 0
    data['emotion'][i] = int(data['emotion'][i])
    img = data['pixels'][i].split()
    img = map(int, img)
    img = np.array(img)
    data['pixels'][i] = img
```

```python
def shuffle(df):
    rows = random.sample(df.index, df.shape[0])
    df = df.ix[rows]
    df = df.reset_index(drop=True)
    return df

train = data[data.Usage=='Training']
val = data[data.Usage!='Training']

x_3 = train[train.emotion==3]
x_0 = train[train.emotion!=3]

x_0.loc[x_0['emotion']!=3,'emotion']=0
val.loc[val['emotion']!=3,'emotion']=0

x_3.loc[x_3['emotion']==3,'emotion']=1
val.loc[val['emotion']==3,'emotion']=1

# x_3 = x_3[:5000]
x_0 = x_0[:15000]

x_3 = x_3.reset_index(drop=True)
x_0 = x_0.reset_index(drop=True)

print x_0.shape
print x_3.shape
print val.shape
d = pd.concat([x_0, x_3])
d = d.reset_index(drop=True)

d = shuffle(d)
print d.shape

print d['pixels'].shape[0], d['pixels'][0].shape[0]
d.head(10)
```

```python
#get mouth area

mouth_x = np.ndarray((d.shape[0],14*24))
eye_x = np.ndarray((d.shape[0],10*32))
lcheek_x = np.ndarray((d.shape[0],19*14))
rcheek_x = np.ndarray((d.shape[0],19*14))

y = np.ndarray((d.shape[0],1))
for i in range(d.shape[0]):
    img = d['pixels'][i]
    img = img.reshape(48, 48)
    
    mouth = img[32:46, 12:36]
    eye   = img[14:24, 8:40]
    lcheek = img[25:44, 5:19]
    rcheek = img[25:44, 30:44]
    
    mouth_x[i] = mouth.reshape(-1)
    
    l = np.linalg.norm(mouth_x[i])
    if l == 0:
#         print l
        mouth_x[i][0] = 1
    eye_x[i] = eye.reshape(-1)
    lcheek_x[i] = lcheek.reshape(-1)
    rcheek_x[i] = rcheek.reshape(-1)
    y[i] = d['emotion'][i]

    
    
print mouth_x.shape, eye_x.shape, lcheek_x.shape, rcheek_x.shape
```
